9. Create your own implementation of a function imgconv (f,w) for applying the
convolution on the following:

```python
def mul(x,y):
  sum=0.0
  for i in range(len(x)):
    for j in range(len(x[0])):
      sum+=x[i][j]*y[i][j]
  return sum
```

```python
def convolution(img,filter):
  X,Y=img.shape
  x,y=filter.shape
  new_img=np.zeros((X-x+1,Y-y+1))
  for r in range(0,X-x+1):
    for c in range(0,Y-y+1):
      new_img[r][c]=mul(img[r:r+x,c:c+y],np.transpose(filter))
  return new_img
```

```python
w=np.array([[1,2,1],[2,4,2],[1,2,1]])
f=np.zeros((5,5))
f[1][2]=1.0
f[2][2]=1.0
f[3][2]=1.0
```

```python
convolution(f,w)
```
